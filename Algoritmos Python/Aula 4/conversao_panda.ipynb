```python
#importando bibliotecas
import pandas as pd

def abrir_filmes_disney():
    df_ex1 = pd.read_csv("disney_plus_shows.csv")

    df_ex1 = df_ex1.replace("N/A", None)
    df_ex1.dropna(subset=['imdb_id'], inplace=True)

    #converter colunas especificas
    df_ex1["metascore"] = pd.to_numeric(df_ex1["metascore"], errors="coerce")
    df_ex1["imdb_rating"] = pd.to_numeric(df_ex1["imdb_rating"], errors="coerce")
    df_ex1["imdb_votes"] = pd.to_numeric(df_ex1["imdb_votes"].astype(str).str.replace(',','',regex=False), errors="coerce")
    df_ex1["year"] = pd.to_numeric(df_ex1["year"].astype(str).str.replace('-','',regex=False), errors="coerce")

    return df_ex1


abrir_filmes_disney()
```
