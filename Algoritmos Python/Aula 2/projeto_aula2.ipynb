```python
#Projeto Aula 2 - Arquivo com Atividades

#Listar as atividades do arquivo
def lista_atividade():
  try:
    # Abre (ou cria) o arquivo em modo de adição (append)
    with open("StatusAtividades.txt", "r", encoding="utf-8") as arquivo:
     linhas = [linha.strip() for linha in arquivo.readlines()]
     cont = 0
     for i in range(0, len(linhas), 2):
      cont += 1
      if i+1 < len(linhas):
       print(f"{cont} - Atividade. {linhas[i]} \n Status: {linhas[i+1]}")
  except IOError as e:
    print(f"Houe um erro ao ler arquivo: {e}")

#Adicionar Atividade
def adicionar_atividade():
  #Recebendo nome e status da atividade
  nomeAtividade = input("Digite o nome da Atividade: ").strip()
  statusAtividade = input("Digite o Status da Atividade: ").strip()
                   
  if nomeAtividade and statusAtividade:
    # Abre (ou cria) o arquivo em modo de adição (append)
    with open("StatusAtividades.txt", "a", encoding="utf-8") as arquivo:
        arquivo.write(nomeAtividade + "\n") 
        arquivo.write(statusAtividade + "\n")
        print("Atividade adicionada com sucesso!")
      

#Remover Atividade
def remover_atividade():
  lista_atividade() #exibe as atividades para escolha
  try:
    numEscolhido = int(input("Escolha a atividade digitando o número referente a ela: "))
    indice = (numEscolhido - 1) * 2

    # Abre (ou cria) o arquivo em modo de adição (append)
    with open("StatusAtividades.txt", "r", encoding="utf-8") as arquivo:
       linhas = arquivo.readlines()       
       if 0 <= indice < len(linhas):
        # Remove a atividade
        linhas.pop(indice)
        # Remove o status (que agora está na mesma posição)
        if indice < len(linhas):
            linhas.pop(indice)
   
    with open("StatusAtividades.txt", "w", encoding="utf-8") as arquivo:
       for linha in linhas:
          arquivo.write(linha)

  except IOError as e:
    print(f"Houe um erro ao tentar remover a atividade do arquivo. Erro: {e}")

def concluir_atividade():
    lista_atividade()
    try:
        numEscolhido = int(input("Número da atividade para concluir: "))
        indice = (numEscolhido - 1) * 2

        with open("StatusAtividades.txt", "r", encoding="utf-8") as arquivo:
          linhas = arquivo.readlines()
          if 0 <= indice < len(linhas):
              linhas[indice + 1] = "Concluída\n"
              with open("StatusAtividades.txt", "w", encoding="utf-8") as arquivo:
                  arquivo.writelines(linhas)

              print("Atividade concluída!\n")
          else:
              print("Índice inválido.\n")
    except ValueError:
        print("Entrada inválida.\n")


#Definindo função
def iniciar_projeto():
    entrada = 0
    #definição do Menu
    # 1 - Incluir Atividade
    # 2 - Remover Atividade
    # 3 - Concluir Atividade
    # 4 - Sair
    try:
        while True:
          # Pede ao usuário para digitar a entrada 
          entrada = input("\n\n#### MENU DE OPÇÕES ##### \n" \
          "# 1 - Incluir Atividade  #\n" \
          "# 2 - Remover Atividade  #\n" \
          "# 3 - Concluir Atividade #\n" \
          "# 4 - Exibir Arquivo     #\n" \
          "# 5 - Sair               #\n" \
          "########################## \n\n" \
          "Escolha a opção desejada informando o número: ")

          if(entrada == "5"):
              print("Encerrando execução")
              break
          else:
              if entrada == "1":
                adicionar_atividade()
              elif entrada == "2":
                remover_atividade()
              elif entrada == "3":
                concluir_atividade()
              elif entrada == "4":
                lista_atividade()
              else:
                print("Opção Inválida")
    except IOError as ex:
        print(f"Houve um erro ao iniciar o projeto. Erro: {ex}")

#chamando inicio do projeto
iniciar_projeto()
```
