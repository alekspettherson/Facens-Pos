```python
import os
arquivo = "SistemaVendas.txt"
#Listar as vendas do arquivo
def listar_Vendas():
  global arquivo
  try:
    # Verifica se o arquivo existe para fazer a leitura
   if os.path.exists(arquivo):
    with open(arquivo, "r", encoding="utf-8") as f:
     linhas = [linha.strip() for linha in f.readlines()]
     cont = 0
     for i in range(0, len(linhas), 3):
      cont += 1
      if i+1 < len(linhas):
       print(f"{cont} - Produto: {linhas[i]} \n Qtd Vendida: {linhas[i+1]} \n Preço: {linhas[i+2]}")
   else:
    print("Nenhum arquivo foi encontrado!")
  except IOError as e:
    print(f"Houe um erro ao ler arquivo: {e}")

#Adicionar vendas
def incluir_venda():
  global arquivo
  #Recebendo nome e status da atividade
  nomeProduto = input("Digite o nome do produto: ").strip()
  qtdVenda = input("Digite a quantidade vendida: ").strip()
  preco = input("Digite o preço do produto: ").strip()
                   
  # Abre (ou cria) o arquivo em modo de adição (append)
  with open(arquivo, "a", encoding="utf-8") as f:
    f.write(nomeProduto + "\n") 
    f.write(qtdVenda + "\n")
    f.write(preco + "\n")
    print("Venda processada com sucesso!")
      

#Remover Atividade
def remover_venda():
  listar_Vendas() #exibe as atividades para escolha
  try:
    global arquivo
    numEscolhido = int(input("Escolha a venda que deseja remover: "))
    indice = (numEscolhido - 1) * 2

    # Abre (ou cria) o arquivo em modo de adição (append)
    with open(arquivo, "r", encoding="utf-8") as f:
       linhas = f.readlines()       
       if 0 <= indice < len(linhas):
        # Remove o produto
        linhas.pop(indice)
        # Remove a quantia (que agora está na mesma posição)
        if indice < len(linhas):
            linhas.pop(indice)
        # Remove o preço (que está na mesma posição)
        if indice < len(linhas):
            linhas.pop(indice)
   
    with open(arquivo, "w", encoding="utf-8") as f:
       for linha in linhas:
          f.write(linha)

  except IOError as e:
    print(f"Houe um erro ao tentar remover a atividade do arquivo. Erro: {e}")

def alterar_venda():
    listar_Vendas()
    try:
        global arquivo
        numEscolhido = int(input("Número da venda que deseja alterar: "))
        indice = (numEscolhido - 1) * 2

        qtdVenda = input("Digite a quantidade vendida: ").strip()
        preco = input("Digite o preço do produto: ").strip()

        with open(arquivo, "r", encoding="utf-8") as f:
          linhas = f.readlines()
          if 0 <= indice < len(linhas):
              linhas[indice + 1] = qtdVenda + "\n"
              linhas[indice + 2] = preco + "\n"
              with open(arquivo, "w", encoding="utf-8") as f:
                  f.writelines(linhas)

              print("Venda Alterada!\n")
          else:
              print("Índice inválido.\n")
    except ValueError:
        print("Entrada inválida.\n")

#Definindo o relatório
def exibir_relatorio():
  global arquivo
  try:
    soma_quantia = 0
    soma_venda = 0
    maior_venda = []
    # Verifica se o arquivo existe para fazer a leitura
    if os.path.exists(arquivo):
     with open(arquivo, "r", encoding="utf-8") as f:
      linhas = [linha.strip() for linha in f.readlines()]
      
      #loop para interagir com a lista 
      for i in range(0, len(linhas), 3):
        soma_quantia += int(linhas[i+1])
        soma_venda += float(linhas[i+2].replace(",", "."))
        produto = linhas[i]
        quantidade = int(linhas[i+1])

        #verificando produto mais vendido
        if not maior_venda or quantidade > maior_venda["quantidade"]:
          maior_venda = {
              "produto": produto,
              "quantidade": quantidade
          }

     valorMedio = soma_venda / soma_quantia

     relatorio = "\n\n#### RELATÓRIO DE VENDAS ####################### \n" \
                f"# Total Itens Vendidos: {soma_quantia}                     #\n" \
                f"# Total das vendas: {soma_venda}                        #\n" \
                f"# Valor médio das vendas: {valorMedio}                  #\n" \
                f"# Produto mais vendido: {maior_venda['produto']} ({maior_venda['quantidade']} unidades) #\n" \
                "################################################ \n\n"
     print(relatorio)
    else:
      print("Nenhum arquivo foi encontrado!")
  except IOError as e:
    print(f"Houe um erro ao ler arquivo: {e}")

#Definindo função
def iniciar_projeto():
    entrada = 0
    #definição do Menu
    # 1 - Incluir Atividade
    # 2 - Remover Atividade
    # 3 - Concluir Atividade
    # 4 - Sair
    try:
        while True:
          # Pede ao usuário para digitar a entrada 
          entrada = input("\n\n#### MENU DE OPÇÕES ##### \n" \
                              "# 1 - Incluir venda     #\n" \
                              "# 2 - Alterar venda     #\n" \
                              "# 3 - Remover venda     #\n" \
                              "# 4 - Relatorio de venda#\n" \
                              "# 5 - Sair              #\n" \
                              "######################### \n\n" \
          "Escolha a opção desejada informando o número: ")

          if(entrada == "5"):
              print("Encerrando execução")
              break
          else:
              if entrada == "1":
               incluir_venda()
              elif entrada == "2":
                remover_venda()
              elif entrada == "3":
                alterar_venda()
              elif entrada == "4":
                exibir_relatorio()
                break
              else:
                print("Opção Inválida")
    except IOError as ex:
        print(f"Houve um erro ao iniciar o projeto. Erro: {ex}")

#chamando inicio do projeto
iniciar_projeto()
```
